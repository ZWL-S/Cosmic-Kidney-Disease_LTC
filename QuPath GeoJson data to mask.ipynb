```python
import json
from shapely.geometry import Polygon
import matplotlib.pyplot as plt

# Load your GeoJSON file
with open(r'let_7a/GEOJSON/F162.geojson', 'r') as file:
    geojson_data = json.load(file)

print(len(geojson_data['features']))
# Extract the coordinates for the polygon and its holes
coordinates = geojson_data['features'][5]['geometry']['coordinates'][0]

# The first set of coordinates is the outer boundary
outer_boundary = coordinates[0]

# The rest of the items in the list are holes (if they exist)
holes = coordinates[1:] if len(coordinates) > 1 else []

# Create a polygon with holes using Shapely
polygon_with_holes = Polygon(shell=outer_boundary, holes=holes)

# Plotting
fig, ax = plt.subplots()
x, y = polygon_with_holes.exterior.xy
ax.fill(x, y, alpha=0.5, fc='blue', ec='none')  # Filling the outer boundary

# Plotting holes
for hole in polygon_with_holes.interiors:
    x, y = hole.xy
    ax.fill(x, y, alpha=1, fc='white', ec='none')  # Filling holes

ax.set_title('Polygon with Holes')
plt.show()
```
